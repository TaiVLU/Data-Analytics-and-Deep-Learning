# BÀI THỰC HÀNH THAO TÁC DỮ LIỆU

```python
import pandas as pd
import numpy as np
```

```python
# 1. Tải dữ liệu lên
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)
```

```python
# 2. Phân loại dữ liệu
qualitative_vars = ['GT', 'DT', 'KV', 'KT']  
quantitative_vars = [col for col in df.columns if col not in qualitative_vars]  
```

```python
# 3. Hiển thị 10 dòng đầu và 10 dòng cuối
display(df.head(10))
display(df.tail(10))
```

```python
# 4. Kiểm tra và xử lý dữ liệu thiếu cho cột DT
df['DT'].fillna(0, inplace=True)
```

```python
# 5. Kiểm tra và xử lý dữ liệu thiếu cho cột T1
mean_T1 = df['T1'].mean()
df['T1'].fillna(mean_T1, inplace=True)
```

```python
# 6. Xử lý dữ liệu thiếu cho các cột điểm số còn lại
score_cols = [col for col in df.columns if col.startswith(('T', 'L', 'H', 'S', 'V', 'X', 'D', 'N', 'DH'))]
df[score_cols] = df[score_cols].apply(lambda x: x.fillna(x.mean()))
```

```python
# 7. Tạo biến TBM1, TBM2, TBM3
def calculate_tbm(row, t, l, h, s, v, x, d, n):
    return (row[t]*2 + row[l] + row[h] + row[s] + row[v]*2 + row[x] + row[d] + row[n]) / 10

df['TBM1'] = df.apply(calculate_tbm, axis=1, args=('T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1'))
df['TBM2'] = df.apply(calculate_tbm, axis=1, args=('T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2'))
df['TBM3'] = df.apply(calculate_tbm, axis=1, args=('T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6'))
```

```python
# 8. Tạo biến xếp loại
bins = [0, 5, 6.5, 8, 9, 10]
labels = ['Y', 'TB', 'K', 'G', 'XS']
df['XL1'] = pd.cut(df['TBM1'], bins=bins, labels=labels, right=False)
df['XL2'] = pd.cut(df['TBM2'], bins=bins, labels=labels, right=False)
df['XL3'] = pd.cut(df['TBM3'], bins=bins, labels=labels, right=False)
```

```python
# 9. Chuyển đổi điểm sang thang điểm 4 (Min-Max Normalization)
def min_max_scaling(col):
    return (col - col.min()) / (col.max() - col.min()) * 4

df['US_TBM1'] = min_max_scaling(df['TBM1'])
df['US_TBM2'] = min_max_scaling(df['TBM2'])
df['US_TBM3'] = min_max_scaling(df['TBM3'])
```

```python
# 10. Tạo biến KQXT
def determine_result(row):
    if row['KT'] in ['A', 'A1']:
        return 1 if (row['DH1']*2 + row['DH2'] + row['DH3']) / 4 >= 5 else 0
    elif row['KT'] == 'B':
        return 1 if (row['DH1'] + row['DH2']*2 + row['DH3']) / 4 >= 5 else 0
    else:
        return 1 if (row['DH1'] + row['DH2'] + row['DH3']) / 3 >= 5 else 0

df['KQXT'] = df.apply(determine_result, axis=1)
```

```python
# 11. Lưu trữ dữ liệu xuống file mới
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
