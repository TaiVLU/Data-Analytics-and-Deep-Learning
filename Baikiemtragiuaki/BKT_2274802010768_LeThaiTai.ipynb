# 14. Biết công thức tính chỉ số BMI như sau:
## BMI = Weight(kg)/Height2 (m2)
### Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người

```python
# Thêm cột cs_BMI để tính chỉ số BMI
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
print("DataFrame sau khi thêm cột cs_BMI:")
print(df)
```

# 15. Theo thang đo kết quả tính BMI của châu Á:
## Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình thường, 23-24,9 là thừa cân, từ 25 trở lên là béo phì.
### Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ số BMI trên.


```python
# Thêm cột Phan_Loai theo tiêu chuẩn BMI của châu Á
def classify_bmi(bmi):
    if bmi < 18.5:
        return '0'  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return '1'  # Bình thường
    elif 23 <= bmi <= 24.9:
        return '2'  # Thừa cân
    else:
        return '3'  # Béo phì

df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
print("DataFrame sau khi thêm cột Phan_Loai:")
print(df)
```

# 16. Chuẩn hóa giá trị của cột cs_BMI chuẩn hóa giá trị trong đoạn từ 0 tới 1

```python
# Chuẩn hóa giá trị cột cs_BMI trong khoảng từ 0 đến 1
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print("DataFrame sau khi chuẩn hóa cột cs_BMI:")
print(df)
```

# 17. Cho biết:
## + ‘Mr‘tương ứng với ‘male’,
## + ’Mrs’ và ‘Miss’ tương ứng với ‘female’. 
### Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": [
        "Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William",
        "Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite",
        "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"
    ],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)

def get_gender(name):
    if name.startswith("Mr."):
        return "male"
    elif name.startswith("Mrs.") or name.startswith("Miss."):
        return "female"
    return "unknown"

df["GT"] = df["Name"].apply(get_gender)
df["Name"] = df["Name"].str.replace(r"^(Mr\.|Mrs\.|Miss\.)\s*", "", regex=True)

print(df)
```

# 18. Vẽ biểu đồ tròn cho GT.

```python
gt_counts = df["GT"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gt_counts, labels=gt_counts.index, autopct='%1.1f%%', colors=["blue", "pink"], startangle=90)
plt.title("Biểu đồ tròn cho GT")
plt.show()
```

```python
if "cs_BMI" not in df.columns:
    df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)

if "Phan_Loai" not in df.columns:
    def classify_bmi(bmi):
        if bmi < 18.5:
            return 0  # Thiếu cân
        elif 18.5 <= bmi <= 22.9:
            return 1  # Bình thường
        elif 23 <= bmi <= 24.9:
            return 2  # Thừa cân
        else:
            return 3  # Béo phì

    df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)

df_female = df[df["GT"] == "female"]

female_counts = df_female["Phan_Loai"].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(female_counts.index, female_counts.values, color="pink")
plt.xlabel("Phân loại BMI")
plt.ylabel("Số lượng")
plt.title("Trực quan hóa dữ liệu người nữ trên Phan_Loai")
plt.xticks([0, 1, 2, 3], ["Thiếu cân", "Bình thường", "Thừa cân", "Béo phì"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
