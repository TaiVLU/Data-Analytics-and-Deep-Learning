# Phần 1: NumPy
1.1: Tạo mảng NumPy

```python
import numpy as np

np.set_printoptions(precision=2, suppress=True)

# Tạo một mảng 1D
array_1d = np.array([1, 2, 3, 4, 5], dtype=int)
print("Mảng 1D:\n", array_1d, "\n")

# Tạo một mảng 2D
array_2d = np.array([[1, 2, 3], [4, 5, 6]], dtype=int)
print("Mảng 2D:\n", array_2d, "\n")

# Tạo mảng số ngẫu nhiên
random_array = np.random.rand(3, 3)
print("Mảng số ngẫu nhiên:\n", random_array)
```

1.2: Các thao tác cơ bản

```python
import numpy as np

array = np.array([10, 20, 30, 40, 50], dtype=int)

# Cộng, trừ, nhân, chia với một số
print("Cộng thêm 10:\n", array + 10, "\n")
print("Nhân với 2:\n", array * 2, "\n")

# Tính tổng, trung bình, lớn nhất, nhỏ nhất
print("Tổng:", array.sum())
print("Trung bình:", array.mean())
print("Giá trị lớn nhất:", array.max())
print("Giá trị nhỏ nhất:", array.min(), "\n")

# Lấy phần tử và lát cắt
print("Phần tử đầu tiên:", array[0])
print("Lát cắt [1:4]:", array[1:4])
```

# Phần 2: Pandas
2.1: Tạo DataFrame

```python
import pandas as pd

# Tạo DataFrame từ dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df.to_string())
```

2.2: Các thao tác cơ bản với DataFrame

```python
import pandas as pd

# Tạo DataFrame từ dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}

df = pd.DataFrame(data)

# Truy cập dữ liệu
print("Cột Age:\n", df["Age"], "\n")
print("Hàng đầu tiên:\n", df.iloc[0], "\n")

# Lọc dữ liệu
filtered_df = df[df["Score"] > 85]
print("Học sinh có điểm > 85:\n", filtered_df.to_string(), "\n")

# Tính toán
print("Tuổi trung bình:", df["Age"].mean())
```

# Phần 3: Matplotlib
3.1: Biểu đồ đường

```python
import matplotlib.pyplot as plt

# Dữ liệu
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.figure(figsize=(6, 4))

# Vẽ biểu đồ đường
plt.plot(x, y, label="Biểu đồ đường", color="blue", marker="o", linestyle="-", linewidth=2, markersize=6)

# Thêm tiêu đề và nhãn
plt.title("Biểu đồ Đường Cơ Bản", fontsize=14)
plt.xlabel("Trục X", fontsize=12)
plt.ylabel("Trục Y", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10)
plt.show()
```

3.2: Biểu đồ cột

```python
import matplotlib.pyplot as plt

# Dữ liệu
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]
plt.figure(figsize=(6, 4))

# Vẽ biểu đồ cột
bars = plt.bar(categories, values, color="skyblue", edgecolor="black")

plt.title("Biểu đồ Cột", fontsize=14)
plt.xlabel("Danh mục", fontsize=12)
plt.ylabel("Giá trị", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(bar.get_height()), 
             ha="center", va="bottom", fontsize=10)

plt.show()
```

3.3: Biểu đồ phân tán

```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu ngẫu nhiên
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 200 
colors = np.arange(50)  

# Tạo biểu đồ với kích thước
plt.figure(figsize=(6, 4))

# Vẽ biểu đồ phân tán
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.7, edgecolors="black")

plt.colorbar(scatter, label="Mức độ màu")
plt.title("Biểu đồ Phân Tán", fontsize=14)
plt.xlabel("Trục X", fontsize=12)
plt.ylabel("Trục Y", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)


plt.show()
```
